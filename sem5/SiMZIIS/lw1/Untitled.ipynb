```python
from password import Password

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
password1: Password = Password(8)
password2: Password = Password(30)
password2.password
```

```python
def counts(password: Password) -> dict:
    counts_dict: dict = {}

    password_str: str = password.password
    for i in range(len(password_str)):
        if password_str[i] not in counts_dict:
            counts_dict[password_str[i]] = 0

        counts_dict[password_str[i]] += 1

    return counts_dict

password1_counts: dict = counts(password1)
password2_counts: dict = counts(password2)
```

```python
password1_data: list[dict] = [{'symbol': key, 'counts': value} for key, value in password1_counts.items()]
password1_dataset: pd.DataFrame = pd.DataFrame(password1_data)

password2_data: list[dict] = [{'symbol': key, 'counts': value} for key, value in password2_counts.items()]
password2_dataset: pd.DataFrame = pd.DataFrame(password2_data)
```

```python
plt.bar(password1_counts.keys(), password1_counts.values(), color='blue')
plt.title('Распределение символов в пароле')
plt.xlabel('Символ')
plt.ylabel('Частота')
plt.show()
```

```python
plt.bar(password2_counts.keys(), password2_counts.values(), color='blue')
plt.title('Распределение символов в пароле')
plt.xlabel('Символ')
plt.ylabel('Частота')
plt.show()
```

```python
def hack_data(n: int) -> dict:
    hack_data_dict: dict = {i: 59**i / 1000000 for i in range(1, n + 1)}
    return hack_data_dict

hack_dict1: dict = hack_data(5)
hack_dict2: dict = hack_data(6)
```

```python
plt.plot(hack_dict2.keys(), hack_dict2.values(), color='blue')
plt.title('Зависимость времени подбора пароля от длины символов')
plt.xlabel('Длина пароля, символы')
plt.ylabel('Время подбора, секунды')
plt.show()
```
